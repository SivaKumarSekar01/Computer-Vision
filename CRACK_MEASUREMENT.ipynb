```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def measure_crack(image_path):
    # Read the image
    image = cv2.imread(image_path)

    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Perform edge detection
    edges = cv2.Canny(blurred, 50, 150)

    # Find contours in the edge-detected image
    contours, _ = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create a mask image with the same dimensions as the original image
    mask = np.zeros_like(image)

    # Fill the mask with a light color inside the crack contour
    cv2.drawContours(mask, contours, -1, (255, 255, 0), -1)

    # Blend the original image and the mask to highlight the crack
    highlighted_image = cv2.addWeighted(image, 0.7, mask, 0.3, 0)

    # Get the coordinates of the pixels in the segmented crack zone
    crack_coordinates = np.column_stack(np.where(mask[:, :, 0] > 0))

    # Check if crack_coordinates is empty
    if crack_coordinates.size == 0:
        print("No crack detected in the image.")
        length_pixels = 0
        width_pixels = 0
    else:
        # Calculate the length and width of the segmented crack zone
        min_x, min_y = np.min(crack_coordinates, axis=0)
        max_x, max_y = np.max(crack_coordinates, axis=0)
        length_pixels = max_x - min_x
        width_pixels = max_y - min_y

        # Draw a line along the segmented crack if crack_coordinates are within bounds
        if max_x < mask.shape[0] and max_y < mask.shape[1]:
            crack_mask = np.zeros_like(mask[:, :, 0])
            crack_mask[crack_coordinates[:, 0], crack_coordinates[:, 1]] = 255
            kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))
            crack_mask = cv2.morphologyEx(crack_mask, cv2.MORPH_DILATE, kernel, iterations=1)
            contours, _ = cv2.findContours(crack_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
            if contours:
                contour = max(contours, key=cv2.contourArea)
                cv2.polylines(highlighted_image, [contour], False, (0, 0, 255), 2)

    # Display the length and width on the image
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(highlighted_image, f'Length: {length_pixels} pixels', (10, 30), font, 0.5, (0, 0, 255), 1, cv2.LINE_AA)
    cv2.putText(highlighted_image, f'Width: {width_pixels} pixels', (10, 60), font, 0.5, (0, 0, 255), 1, cv2.LINE_AA)

    # Display the image with the crack measurement, segmentation map, and line
    cv2_imshow(highlighted_image)

    return length_pixels, width_pixels

# Example usage
image_path = "/content/download.jpg"  # Replace with the path to your image
length, width = measure_crack(image_path)
print("Length of the crack:", length, "pixels")
print("Width of the crack:", width, "pixels")
```
